```python
from utils import pr
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,40,1000)
pr(t[:12] , 'time')
```

```python
# distance traveled by the robber
d_r = 2.5 * t
d_a = 3 * (t - 5)
d_r[:10]  , d_a[:10]
```

```python
fig , ax = plt.subplots()
ax.set_title('A bank robber caught')
ax.set_xlabel('time (in minutes)')
ax.set_ylabel('distance (in KM)')
ax.set_xlim((0,40))
ax.set_ylim((0,100))
ax.plot(t , d_r , c='green')
ax.plot(t, d_a , c='brown')
ax.axvline(x=30 , c='purple' , linestyle='--')
ax.axhline(y=75 , c='purple' , linestyle='--')
fig.tight_layout()
```
