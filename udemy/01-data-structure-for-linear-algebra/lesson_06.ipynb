```python
X = np.array([[25, 2], 
              [5, 26], 
              [3, 7]])
pr( X.shape, 'X.shape')
pr( X.size, 'X.size')
pr( X[:,0], 'X[:,0]')
pr( X[1,:], 'X[1,:]')
pr( X[0:2,0:2], 'X[0:2,0:2]')
```

```python
X = torch.tensor([[25, 2], 
              [5, 26], 
              [3, 7]])
pr( X.shape, 'X.shape')
pr( X.size, 'X.size')
pr( X[:,0], 'X[:,0]')
pr( X[1,:], 'X[1,:]')
pr( X[0:2,0:2], 'X[0:2,0:2]')
```
