```python
from utils import pr , measure_time
import numpy as np
```

```python
v = np.array([2.0, 0.0])
u = np.array([1.0, 0.0])
```

```python
dot_product = np.dot(v,u)
pr(dot_product , 'dot_product')
norm_v = np.linalg.norm(v)
norm_u = np.linalg.norm(u)
pr( norm_u , 'norm_u')
pr( norm_v , 'norm_v')
cos_similarity = dot_product / (norm_u * norm_v)
pr(cos_similarity , 'cos_similarity')
```
